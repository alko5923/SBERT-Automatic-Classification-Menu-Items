Firstly, let us load the prediction dataframes from csv files.

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report

# Define the categories
categories = ['Beer', 'Burger', 'Champagne', 'Cider', 'CoffeeTeaMilk', 'Dessert', 
              'Kids', 'Other', 'Salad', 'SoftDrinks', 'Spirits', 'VariedFood', 'Wine', 'Wings']
```

```python
# Load the prediction dataframes
df_predict_distil_orig = pd.read_csv('/content/df-predict-distil-orig.csv')
df_predict_distil_proc = pd.read_csv('/content/df-predict-distil-proc.csv')
df_predict_para_mini_orig = pd.read_csv('/content/df-predict-para-mini-orig.csv')
df_predict_para_mini_proc = pd.read_csv('/content/df-predict-para-mini-proc.csv')
df_predict_para_base_orig = pd.read_csv('/content/df-predict-para-base-orig.csv')
df_predict_para_base_proc = pd.read_csv('/content/df-predict-para-base-proc.csv')
```

```python
# Print the accuracy results for the distiluse model
# Original data
y_truth_distil_orig = df_predict_distil_orig['truth'].values.tolist()
y_predict_distil_orig = df_predict_distil_orig['prediction'].values.tolist()
# Processed data
y_truth_distil_proc = df_predict_distil_proc['truth'].values.tolist()
y_predict_distil_proc = df_predict_distil_proc['prediction'].values.tolist()

# Create the classification reports
class_report_distil_orig = classification_report(y_truth_distil_orig, 
                                                 y_predict_distil_orig, output_dict=True)
class_report_distil_proc = classification_report(y_truth_distil_proc, 
                                                 y_predict_distil_proc, output_dict=True)
# Create the confusion matrix
cm_distil_orig = ConfusionMatrixDisplay.from_predictions(y_truth_distil_orig, 
                                                         y_predict_distil_orig, xticks_rotation='vertical')
cm_distil_proc = ConfusionMatrixDisplay.from_predictions(y_truth_distil_proc,
                                                         y_predict_distil_proc, xticks_rotation='vertical')

# Save the reports as CSV files
df_report_distil_orig = pd.DataFrame(class_report_distil_orig).transpose()
df_report_distil_proc = pd.DataFrame(class_report_distil_proc).transpose()
df_report_distil_orig.to_csv('df-report-distil-orig.csv')
df_report_distil_proc.to_csv('df-report-distil-proc.csv')

print("The distil-orig report \n", class_report_distil_orig)
print("The distil-proc report \n", class_report_distil_proc)
```

```python
# Print the accuracy results for the para-mini model
# Original data
y_truth_para_mini_orig = df_predict_para_mini_orig['truth'].values.tolist()
y_predict_para_mini_orig = df_predict_para_mini_orig['prediction'].values.tolist()
# Processed data
y_truth_para_mini_proc = df_predict_para_mini_proc['truth'].values.tolist()
y_predict_para_mini_proc = df_predict_para_mini_proc['prediction'].values.tolist()
# Create the classification reports
class_report_para_mini_orig = classification_report(y_truth_para_mini_orig,
                                                    y_predict_para_mini_orig, output_dict=True)
class_report_para_mini_proc = classification_report(y_truth_para_mini_proc,
                                                    y_predict_para_mini_proc, output_dict=True)
# Create the confusion matrix
cm_para_mini_orig = ConfusionMatrixDisplay.from_predictions(y_truth_para_mini_orig, y_predict_para_mini_orig, xticks_rotation='vertical')
cm_para_mini_proc = ConfusionMatrixDisplay.from_predictions(y_truth_para_mini_proc, y_predict_para_mini_proc, xticks_rotation='vertical')

# Save the reports as CSV files
df_report_para_mini_orig = pd.DataFrame(class_report_para_mini_orig).transpose()
df_report_para_mini_proc = pd.DataFrame(class_report_para_mini_proc).transpose()
df_report_para_mini_orig.to_csv('df-report-para-mini-orig.csv')
df_report_para_mini_proc.to_csv('df-report-para-mini-proc.csv')

print(class_report_para_mini_orig)
print(class_report_para_mini_proc)
```

```python
# Print the accuracy results for the para-base model
# Original data
y_truth_para_base_orig = df_predict_para_base_orig['truth'].values.tolist()
y_predict_para_base_orig = df_predict_para_base_orig['prediction'].values.tolist()
# Processed data
y_truth_para_base_proc = df_predict_para_base_proc['truth'].values.tolist()
y_predict_para_base_proc = df_predict_para_base_proc['prediction'].values.tolist()
# Create the classification reports
class_report_para_base_orig = classification_report(y_truth_para_base_orig,
                                                    y_predict_para_base_orig, output_dict=True)
class_report_para_base_proc = classification_report(y_truth_para_base_proc, 
                                                    y_predict_para_base_proc, output_dict=True)
# Create the confusion matrix
cm_para_base_orig = ConfusionMatrixDisplay.from_predictions(y_truth_para_base_orig, y_predict_para_base_orig, xticks_rotation='vertical')
cm_para_base_proc = ConfusionMatrixDisplay.from_predictions(y_truth_para_base_proc, y_predict_para_base_proc, xticks_rotation='vertical')

# Save the reports as CSV files
df_report_para_base_orig = pd.DataFrame(class_report_para_base_orig).transpose()
df_report_para_base_proc = pd.DataFrame(class_report_para_base_proc).transpose()
df_report_para_base_orig.to_csv('df-report-para-base-orig.csv')
df_report_para_base_proc.to_csv('df-report-para-base-proc.csv')

print(class_report_para_base_orig)
print(class_report_para_base_proc)
```

We will now visualise the different scores for all three models.

```python
''' Visualise the confusion matrices for all datasets. 
Models: 
'distiluse-base-multilingual-cased-v2',  
'paraphrase-multilingual-MiniLM-L12-v2' and
'paraphrase-multilingual-mpnet-base-v2'. 
'''
fig = plt.figure()
fig.set_size_inches(9,9)
ax = fig.add_subplot(111)

# Subplots - change here for other datasets 
cm_para_base_proc.plot(xticks_rotation='vertical', ax=ax)

# Set the suptitle and common labels
fig.suptitle('Confusion matrix for the paraphrase-multilingual-mpnet-base-v2 model, processed data')

# Change here for different datasets
plt.savefig('cm-para-base-proc', dpi=300)
plt.show()
```

```python
'''
Visualise the F1 score per category for all datasets.
Models: 
'distiluse-base-multilingual-cased-v2',  
'paraphrase-multilingual-MiniLM-L12-v2' and
'paraphrase-multilingual-mpnet-base-v2'. 
'''

# Plot grid
fig = plt.figure()
fig.set_size_inches(8,8)
ax = fig.add_subplot(111)

ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

# Turn off axis lines and ticks of the big subplot
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(labelcolor='w', top=False, bottom=False, left=False, right=False)

f1_distil_proc = df_report_distil_proc['f1-score'].values.tolist()
f1_para_mini_proc = df_report_para_mini_proc['f1-score'].values.tolist()
f1_para_base_proc = df_report_para_base_proc['f1-score'].values.tolist()

# Subplots - change arguments here for every score
ax1.bar(categories, f1_distil_proc[:14])
ax2.bar(categories, f1_para_mini_proc[:14])
ax3.bar(categories, f1_para_base_proc[:14])

ax1.tick_params(axis='x',labelrotation=45)
ax2.tick_params(axis='x',labelrotation=45)
ax3.tick_params(axis='x',labelrotation=45)

# Set the suptitle and common labels
fig.suptitle('F1 score per category for the 3 models, processed data')
ax.set_ylabel('F1 score')

# Set the subplot titles
ax1.set_title('distiluse-base-multilingual-cased-v2')
ax2.set_title('paraphrase-multilingual-MiniLM-L12-v2')
ax3.set_title('paraphrase-multilingual-mpnet-base-v2')

# Adjust sizes of subplots automatically
plt.tight_layout()

# Adjust the suptitle overlap
plt.subplots_adjust(top=0.9)

# Save and plot - change here for every score
plt.savefig('f1-cat-proc', dpi=300)
plt.show()
```
